## 激活函数
多层感知机，就是从线性转到非线性，在线性的基础上加上非线性激活函数，通常使用的是sigmoid，relu以及tanh函数。     
主要看其导数形式，tanh和sigmoid的导数很像，但是sigmoid的导数最高上线0.25，因此在反向传播过程中如果层数过多，容易导致梯度消失。

```python
import torch
%matplotlib inline
import matplotlib.pyplot as plt
```

### sigmoid函数
因为导数在0.25，容易引起梯度消失

```python
x=torch.arange(-8,8,0.1,requires_grad=True)
y=torch.sigmoid(x)
plt.plot(x.detach(),y.detach(),'-')
plt.ylabel('sigmoid(x)')
```

```python
### 其导数的图像
# x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad,'--')
plt.ylabel('grad of sigmoid')
```

### Relu函数

```python
y=torch.relu(x)
plt.plot(x.detach(),y.detach(),'-')
plt.ylabel('relu(x)')
```

```python
### 其导数的图像
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad,'--')
plt.ylabel('grad of relu')
```

### tanh函数

```python
y=torch.tanh(x)
plt.plot(x.detach(),y.detach(),'-')
plt.ylabel('tanh(x)')
```

```python
### 其导数的图像
x.grad.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad,'--')
plt.ylabel('grad of tanh')
```

## 小结

* 多层感知机在输出层和输入层之间增加一个或多个全连接隐藏层，并通过激活函数转换隐藏层的输出。
* 常用的激活函数包括ReLU函数、sigmoid函数和tanh函数。
